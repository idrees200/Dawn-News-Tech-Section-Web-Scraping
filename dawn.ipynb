```python
result = requests.get(URL)
content = result.text
soup = BeautifulSoup(content, 'html.parser')
print(soup.prettify())
```

```python
boc=soup.find('span',class_='story__time')
time=soup.find('span',class_='timestamp__time').get_text()
date=soup.find('span',class_='timestamp__calendar').get_text()
print(boc)

#timedate
print(time)
print(date)
```

```python
title_D=[]
time_D=[]
date_D=[]
link_D=[]
media_D=[]
exerpet_D=[]
```

```python
box=soup.find_all('article',class_='story')

for b in box:

    title_element = b.find("a", class_="story__link").get_text()
    time_element=b.find('span',class_='timestamp__time').get_text()
    date_element=b.find('span',class_='timestamp__calendar').get_text()
    bos=b.find('div',class_='story__excerpt').get_text()
    storymedia=elements.find('a')
    storylinki=b.find('a',class_='story__link')
    title_D.append(title_element)
    time_D.append(time_element)
    date_D.append(date_element)
    link_D.append(storylinki)
    media_D.append(storymedia)
    exerpet_D.append(bos)
    print(title_element,end="\n"*2)
    print(time_element,end="\n"*2)
    print(date_element,end="\n"*2)
    print(storylinki['href'],end="\n"*2)
    print(storymedia,end="\n"*2)
    print(bos,end="\n"*2)

link_D.append(b.find('a',class_='story__link'))
```

```python
import pandas as pd
print(len(title_D))
print(len(date_D))
print(len(time_D))
print(len(link_D))
print(len(media_D))
print(len(exerpet_D))
df = pd.DataFrame({'title': title_D, 'time': time_D,'date ':date_D,'link ':link_D,'excerpet ':exerpet_D,'media ': media_D})
df.to_csv('21L-5632_q2.csv',encoding='utf-8-sig')
```
